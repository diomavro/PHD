```python
import matplotlib
matplotlib.use('Qt5Agg')

from pylab import *
from matplotlib.animation import FuncAnimation
import matplotlib.pyplot as plt
import networkx as nx
import random as rd

threshold = 5

def init():
    global g, positions, s
    g = nx.karate_club_graph()
    for i in g:
        if random() < .5:
            g.node[i]['state'] = 1 
            g.node[i]['bias'] = 1
        else:
            g.node[i]['state'] = 0
            g.node[i]['bias'] = -1
        if random() < .5:
            g.node[i]['food'] = 1 
            g.node[i]['salty'] = 1
        else:
            g.node[i]['food'] = 0
            g.node[i]['salty'] = -1
    positions = nx.spring_layout(g)
    
def update():
    global g, labels
    speaker = rd.choice(list(g.nodes()))
    listener = rd.choice(list(g.neighbors(speaker))) 
    if random() < .5:
        update_state(speaker, listener)
    else: 
        update_food(speaker, listener)           
    k = g.neighbors(listener) #### Changes!
    for j in k:#### Changes!
        if g.node[j]['state'] != g.node[listener]['state'] and g.node[j]['food'] != g.node[listener]['food']: #### Changes!
            if abs(g.node[j]['bias']) == threshold and abs(g.node[j]['salty']) == threshold: #### Changes!
                g.remove_edge(listener, j) #### Changes!
                
    labels={}
    for i in g:
        labels[i]= g.node[i]['bias']

def update_state(speaker, listener):
    global g
    if abs( g.node[listener]['bias'] ) < threshold  :
        if g.node[listener]['state'] == g.node[speaker]['state']:
            if g.node[listener]['state'] == 1:
                g.node[listener]['bias'] = g.node[listener]['bias'] + 1
            else:
                g.node[listener]['bias'] = g.node[listener]['bias'] - 1
        elif g.node[listener]['state'] == 1 and g.node[listener]['bias'] != 0:
            g.node[listener]['bias'] = g.node[listener]['bias'] - 1
        elif g.node[listener]['state'] == 0 and g.node[listener]['bias'] != 0:
            g.node[listener]['bias'] = g.node[listener]['bias'] + 1            
            #The bias add on depends on the difference between them
        else:
            g.node[listener]['state'] = g.node[speaker]['state']

def update_food(speaker, listener):
    ## update food
    global g 
    if abs( g.node[listener]['salty'] ) < threshold  :#### Changes!
        if g.node[listener]['food'] == g.node[speaker]['food']:
            if g.node[listener]['food'] == 1:
                g.node[listener]['salty'] = g.node[listener]['salty'] + 1
            else:
                g.node[listener]['salty'] = g.node[listener]['salty'] - 1
        elif g.node[listener]['food'] == 1 and g.node[listener]['salty'] != 0:
            g.node[listener]['salty'] = g.node[listener]['salty'] - 1
        elif g.node[listener]['food'] == 0 and g.node[listener]['salty'] != 0:
            g.node[listener]['salty'] = g.node[listener]['salty'] + 1            
            #The bias add on depends on the difference between them
        else:
            g.node[listener]['food'] = g.node[speaker]['food']
            
def init_fig():
    global fig, viz
    fig = figure()
    viz = nx.draw(g, with_labels=False, pos = positions,
                  node_size = [300 + 15*g.nodes[t]['salty'] for t in g.nodes ],
                  node_color = [choose_color(n,
                                            g.nodes[n]['state'],  g.nodes[n]['bias'],
                                            g.nodes[n]['food'],g.nodes[n]['salty']) for n in g.nodes])

def choose_color(node_num, state, bias,food,salty):
    if state == 0 and food == 0:
        if abs(bias) < threshold and abs(salty) < threshold:
            return plt.cm.Purples(0.5)
        else:
            return plt.cm.Purples(0.9)
    elif state == 0 and food == 1:
        if abs(bias) < threshold and abs(salty) < threshold:
            return plt.cm.Blues(0.5)
        else:
            return plt.cm.Blues(0.9)        
    elif state == 1 and food == 1:
        if abs(bias) < threshold and abs(salty) < threshold:
            return plt.cm.Reds(0.5)
        else:
            return plt.cm.Reds(0.9)
    else:
        if abs(bias) < threshold and abs(salty) < threshold: 
            return plt.cm.Greens(0.5)
        else:
            return plt.cm.Greens(0.9)
    

    
def observe():
    cla()
    global g, labels
    #cla()
    nx.draw(g, cmap = cm.binary, vmin = 0, vmax = 1,
        node_size = [300 + 15*g.nodes[t]['salty'] for t in g.nodes ],
        node_color = [choose_color(n, 
                                   g.nodes[n]['state'],g.nodes[n]['bias'], 
                                   g.nodes[n]['food'],g.nodes[n]['salty'])
                               for n in g.nodes],
            pos = positions)
    nx.draw_networkx_labels(g,positions,labels)
                      
    
########################################
init()
init_fig()
step = 50000000000
def step(t):
    update()
    observe()
    return [viz]
ani = FuncAnimation(fig, step, interval=0)
show()
```
