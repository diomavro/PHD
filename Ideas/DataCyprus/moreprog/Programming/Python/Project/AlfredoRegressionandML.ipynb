```python
#CHeck MLPC classifier
from sklearn.datasets import load_digits
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.ensemble import RandomForestClassifier

#getdata
digits = load_digits()
X = digits.data
y = digits.target
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.15, random_state = 42)
#Build Model
clf = MLPClassifier((100,))
clf_random = RandomForestClassifier()
#Fit
clf.fit(X_train,y_train)
clf_random.fit(X_train,y_train)
#Evaluate, look up accuracy and other shit which I will type in due time.  confussioon matric, precision and recall
print clf.score(X_test, y_test)
y_hat = clf.predict(X_test)
print confusion_matrix(y_test,y_hat)
print classification_report(y_test,y_hat)

print clf.score(X_test,y_hat)
print clf_random.score(X_test,y_hat)
```

```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as pl
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

#get data
X,y = make_blobs(n_samples = 100,n_features = 2, centers = 3,random_state = 6)
#create model
kmeans = KMeans(n_clusters = 3)
#fit + #transform
kmeans.fit(X)

cluster_labels = kmeans.labels_

x1,x2 = zip(*X)
# print X
pl.scatter(x1,x2,c = cluster_labels, s =50,cmap = pl.cm.Set2)
pl.show()

for k in range(2,10):
    kmeans = KMeans(n_clusters = k)
    kmeans.fit(X)
    print k, silhouette_score(X, kmeans.labels_)
```
