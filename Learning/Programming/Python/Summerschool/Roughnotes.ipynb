Continous Time Model

```python
## dx:dt = ax(1-x)-x
%matplotlib inline
a = 4
t = 0
x = 0.1
dt = 0.001
data = [x]
ts = [t]

while t < 10:
    x = (a*x*(1-x)-x)*dt+x
    data.append(x)
    t = t+dt
    ts.append(t)
plot(ts,data)
```

```python
## dx:dt = ax(1-x)-x
%matplotlib inline
a = 100
b=1.1
t = 0
x = 0.1
y = 0.2
xdata = [x]
ydata = [y]
ts = [t]    

def update(xx,yy,aa,bb):
    new_x = aa*xx-xx*yy**2
    new_y = bb*yy*xx**2-yy
    return [new_x, new_y] 

for i in xrange(10000):
    New=update(x,y,a,b)
    
    
    x = New[0]*.001+x
    y = New[1]*.001+y
    xdata.append(x)
    ydata.append(y)  
 
plot(xdata)
plot(ydata,'bo')
```
