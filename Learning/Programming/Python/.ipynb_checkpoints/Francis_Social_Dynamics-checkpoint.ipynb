```python
import matplotlib
matplotlib.use('Qt5Agg')

from pylab import *
from matplotlib.animation import FuncAnimation
import matplotlib.pyplot as plt
import networkx as nx
import random as rd

threshold = 10

def init(): 
    global g, positions, s
    g = nx.karate_club_graph()
    for i in g:
        g.node[i]['state'] = 1 if random() < .5 else 0 #gives nodes that are of different states
        if g.node[i]['state'] == 1: ### Changes 
            g.node[i]['bias'] = 1 #### Changes 
        else: #### Changes!
            g.node[i]['bias'] = -1 #### Changes! 
    positions = nx.spring_layout(g)
    
    
def update():
    global g, labels
    listener = rd.choice(list(g.nodes()))
    speaker = rd.choice(list(g.neighbors(listener)))
    if abs( g.node[listener]['bias'] ) < threshold  :## If a nodes absolute value is less than 10
        if g.node[listener]['state'] == g.node[speaker]['state']: ## If the node's state = chosen speaker's(neighbor's) state
            if g.node[listener]['state'] == 1: ## If it is blue, 
                g.node[listener]['bias'] = g.node[listener]['bias'] + 1 ## Add 1 to it
            else:
                g.node[listener]['bias'] = g.node[listener]['bias'] - 1
        elif g.node[listener]['state'] == 1 and g.node[listener]['bias'] != 0:
            g.node[listener]['bias'] = g.node[listener]['bias'] - 1
        elif g.node[listener]['state'] == 0 and g.node[listener]['bias'] != 0:
            g.node[listener]['bias'] = g.node[listener]['bias'] + 1            
            #The bias add on depends on the difference between them
        else:
            g.node[listener]['state'] = g.node[speaker]['state']
    labels={}
    for i in g:
        labels[i]=g.node[i]['bias']
    k = g.neighbors(listener)  
    for j in k:
        if g.node[j]['state'] != g.node[listener]['state']: 
            if abs(g.node[j]['bias']) == threshold:
                g.remove_edge(listener, j)     
        
def init_fig():
    global fig, viz
    fig = figure()
    viz = nx.draw(g, with_labels=False, pos = positions,
                 node_color = [choose_color(g.nodes[n]['state'], n, g.nodes[n]['bias']) for n in g.nodes])
        

def choose_color(state, node_num, bias):
    if state == 0:
        if abs(bias) < threshold:
            return plt.cm.Blues(0.5)
        else:
            return plt.cm.Blues(0.9)
    elif state == 1:
        if abs(bias) < threshold:
            return plt.cm.Reds(0.5)
        else:
            return plt.cm.Reds(0.9)

 
    
def observe():
    cla()
    global g, labels
    #cla()
    nx.draw(g, cmap = cm.binary, vmin = 0, vmax = 1,
        node_color = [choose_color(g.nodes[n]['state'], n,g.nodes[n]['bias'])
                               for n in g.nodes],
            pos = positions)
    nx.draw_networkx_labels(g,positions,labels)
                      
    
########################################
init()
init_fig()
step = 50000000000
def step(t):
    update()
    observe()
    return [viz]
ani = FuncAnimation(fig, step, interval=0)
show()
```
