```python
#Begin by sorting the data
sheet = sheet.sort_values(by="CREATED_DATE")

#We create two functions that might be useful

#Function to get value of row specified
def return_value(data,row,column):
    return data.loc[row,column]

#A function to get the number of unique value in a column
def unique_values(data, column, rows):
    #Creates a set from the first point of data
    list = {data.loc[0][column]}
    for x in range(rows):
        list.add(data.loc[x][column])
    return list
```

```python
#Create a new column called Month that will take the first 7 elements of the date
file1['Month'] = file1['created_at'].str[5:7]
file2['Month'] = file2['created_at'].str[5:7]


#Transform the new column into integers
pd.to_numeric(file1['Month'] )
pd.to_numeric(file2['Month'] )

#Sort values by
#file1.sort_values(by="Month")

#Group by months
Monthly_group = file1.groupby('Month')
Monthly_group2 = file2.groupby('Month')

#Get unique customers in each group
Monthly_group.agg({"user_id": "nunique"})
Monthly_group2.agg({"user_id": "nunique"})

#Group by customer_ID
ID_group = file1.groupby('user_id')
ID_group2 = file2.groupby('user_id')

#Function to group up customers by month and id to see how often they got checked
def monthly_clearance_rate(data, column):
    data['Dummy'] = np.where(data[column] == 'clear',1,
                    np.where(data[column] == 'caution',0,
                    np.where(data[column] == 'consider',0,
                    np.where(data[column] == '',0,
                    np.where(data[column] == 'suspected',0,
                    np.where(data[column] == 'rejected',0,0))))))
    rate = data.groupby(['user_id','Month']).Dummy.agg('sum')
    return rate

#/data.groupby(['user_id','Month']).size()

#Compute the proportion of entries which are cleared per month
def clear_compute(data, column):
    data['Dummy'] = np.where(data[column] == 'clear',1,
                    np.where(data[column] == 'caution',0,
                    np.where(data[column] == 'consider',0,
                    np.where(data[column] == '',0,
                    np.where(data[column] == 'suspected',0,
                    np.where(data[column] == 'rejected',0,0))))))
    rate = data.groupby('Month').Dummy.agg('sum')/data.groupby('Month').size()
    return rate

#Now instead of dividing by the entries in each group, we divide by the unique customers in each month
#Reason: This should avoid double counting customers who have done it twice. 
def clear_compute_unique(data, column):
    data['Dummy'] = np.where( data[column] == 'clear',1,
                    np.where(data[column] == 'caution',0,
                    np.where(data[column] == 'consider',0,
                    np.where(data[column] == '',0,
                    np.where(data[column] == 'suspected',0,
                    np.where(data[column] == 'rejected',0,0))))))
    rate = data.groupby('Month').Dummy.agg('sum')/data.groupby('Month').agg({"user_id": "nunique"}).squeeze()
    return rate
```

```python
#Compute the clear rate to see the trend
print(clear_compute(file1,'result'))
print(clear_compute(file1,'sub_result'))

print(clear_compute_unique(file1,'result'))
print(clear_compute_unique(file1,'sub_result'))
```
