```python
import numpy as np
import matplotlib.pyplot as plt
import random
import pylab
#%matplotlib qt
#%matplotlib inline
```

```python
data = []
data2 = []
data3 = []

def addparetoobs(vector, size):
    vector = np.append(vector, np.random.pareto(0.5,size))
    return vector

def addbinomialobs(vector, size):
    vector = np.append(vector, np.random.binomial(inputfloat(),inputfloat(),size))
    return vector

def addnormalobs(vector, size):
    vector = np.append(vector, np.random.normal(mu,sigma,size))
    return vector

def adduniformobs(vector,  minimum, maximum, size):
    vector = np.append(vector, np.random.uniform(minimum,maximum,size))
    return vector

def createmovingmean(vector):
    means = []
    for i in range(len(vector)):
        means = np.append(means, np.mean(vector[0:i+1])) 
    return means

lower = 10
upper = 20
mu = 10
sigma = 100
n = 300

data = addnormalobs(data, n)
label1 = "Normal"
random.shuffle(data)

data2 = adduniformobs(data2,lower,upper, n)
label2 = "Uniform"
random.shuffle(data2)

data3 = addparetoobs(data3, n)
label3 = "Pareto"
random.shuffle(data3)

meansv = createmovingmean(data)
meansv2 = createmovingmean(data2)
meansv3 = createmovingmean(data3)


plt.style.use('seaborn-deep')
plt.plot(meansv, label = label1)
plt.plot(meansv2, label = label2)
plt.plot(meansv3, label = label3)
plt.legend(loc='upper right')
plt.title("Mean with sample")


plt.tight_layout()
plt.show()
```

```python
sampleSize = 1000

## Let's simulate the repeated throwing of a single six-sided die
singleDie = []
for i in range(sampleSize):
    newValue = random.randint(1,6)
    singleDie.append(newValue)

print( "Results for throwing a single die", sampleSize, "times.")
print( "Mean of the sample =", pylab.mean(singleDie))
print( "Median of the sample =", pylab.median(singleDie))
print( "Standard deviation of the sample =", pylab.std(singleDie))


pylab.hist(singleDie, bins=[0.5,1.5,2.5,3.5,4.5,5.5,6.5] )
pylab.xlabel('Value')
pylab.ylabel('Count')
pylab.savefig('singleDie.png')
pylab.show()

pylab.hist(singleDie, bins=pylab.arange(1.5,12.6,1.0), density = True, histtype='step', cumulative=True)
pylab.xlabel('Value')
pylab.ylabel('Fraction')
pylab.show()

## What about repeatedly throwing two dice and summing them?
twoDice = []
for i in range(sampleSize):
    newValue = random.randint(1,6) + random.randint(1,6)
    twoDice.append(newValue)

print( "Results for throwing two dices", sampleSize, "times.")
print( "Mean of the sample =", pylab.mean(twoDice))
print( "Median of the sample =", pylab.median(twoDice))
#print "Standard deviation of the sample =", pylab.std(twoDice)

pylab.hist(twoDice, bins= pylab.arange(1.5,12.6,1.0))
pylab.xlabel('Value')
pylab.ylabel('Count')
pylab.savefig('twoDice.png')
pylab.show()

pylab.hist(twoDice, bins=pylab.arange(1.5,12.6,1.0), density = True, histtype='step', cumulative=True)
pylab.xlabel('Value')
pylab.ylabel('Fraction')
pylab.show()
```
