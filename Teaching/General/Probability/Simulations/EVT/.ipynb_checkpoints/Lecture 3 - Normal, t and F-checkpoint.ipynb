```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
from scipy.stats import t

# Plot between -5 and 5 with .001 steps.
x_axis = np.arange(-10, 10, 0.001)
# Mean = 0, SD = 2.
plt.plot(x_axis, norm.pdf(x_axis,1,2), label='1 mean')

# Mean = 1, SD = 2.
plt.plot(x_axis, norm.pdf(x_axis,0,2), label='0 mean')

# Mean = 1, SD = 5
plt.plot(x_axis, norm.pdf(x_axis,0,5), label='Bigger variance')

plt.legend(loc='best', frameon=False)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
from scipy.stats import t

x_axis = np.arange(-10, 10, 0.001)

# The second parameter represents the degrees of freedom, here 1
plt.plot(x_axis, t.pdf(x_axis,1), label='t')

# Degrees of freedom here is 100, approaches normal
plt.plot(x_axis, t.pdf(x_axis,100), label='t2')

plt.legend(loc='best', frameon=False)

plt.show()
```

```python
# F distribution
#https://www.johndcook.com/blog/distributions_scipy/

import numpy as np
from scipy.stats import f, norm
import matplotlib.pyplot as plt

# Plot between -5 and 5 with .001 steps.
x_axis = np.arange(0, 6, 0.001)
#A small degree F distribution
plt.plot(x_axis, f.pdf(x_axis, dfn=3, dfd=3), label='F at 3-3')

#A large numerator but small denominator F distribution
plt.plot(x_axis, f.pdf(x_axis, dfn=30, dfd=3), label='Larger numerator')

#A small numerator but large denominator F distribution
plt.plot(x_axis, f.pdf(x_axis, dfn=3, dfd=30), label='Larger Denominator')

#Large degrees of freedom on both sides
plt.plot(x_axis, f.pdf(x_axis, dfn=100, dfd=30), label='Both larger')

plt.legend(loc='best', frameon=False)

plt.show()
```

```python
# Sample code to do tests on
a = 0.05  # set the cutoff

x = np.random.normal(size=1)

rv1 = f(dfn=3, dfd=15, loc=0, scale=1)
rv2 = f(dfn=10, dfd=50, loc=0, scale=1)

p1 = rv1.sf(x)
if p1 < a:
    print('F1 cutoff at: ', x, p1)
else:
    print('F1 No cutoff', x, p1)
    
p2 = rv2.sf(x)
if p2 < a:
    print('F2 cutoff at: ', x, p2)
else:
    print('F2 No cutoff', x, p2)
```
